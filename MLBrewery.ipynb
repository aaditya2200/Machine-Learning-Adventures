```
import pandas as pd
```

```
data = pd.read_csv('cost_revenue_clean.csv')
```

```
data.describe()
```

```
x = pd.DataFrame(data, columns = ['production_budget_usd'])
y = pd.DataFrame(data, columns = ['worldwide_gross_usd'])
```

```
import matplotlib as mpt
```

```

mpt.pyplot.figure(figsize = (15,9))
mpt.pyplot.xlabel('Budget')
mpt.pyplot.ylabel('Revenue')
mpt.pyplot.title('Budget vs. Revenue For Films.')
mpt.pyplot.ylim(0,3000000000)
mpt.pyplot.xlim(0,400000000)
mpt.pyplot.scatter(x,y, alpha=0.3)
```

```
from sklearn.linear_model import LinearRegression
```

```
regression = LinearRegression() #seems like a constructor
```

```
regression.fit(x,y)
```

Slope Coefficient: 

```
regression.coef_ #theta1
```

```
regression.intercept_ #theta0
```

```
mpt.pyplot.figure(figsize = (15,9))
mpt.pyplot.xlabel('Budget')
mpt.pyplot.ylabel('Revenue')
mpt.pyplot.title('Budget vs. Revenue For Films.')
mpt.pyplot.ylim(0,3000000000)
mpt.pyplot.xlim(0,400000000)
mpt.pyplot.scatter(x,y, alpha=0.3)
mpt.pyplot.plot(x,regression.predict(x),color = 'red', linewidth = 4)
```

```
regression.score(x,y)
```
